```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/warehouse_ops.csv', parse_dates=['date'])
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
```

```python
df['pick_rate'] = df['items_picked'] / df['hours_worked']
df['pick_rate'].hist(bins=40)
plt.title('Pick rate distribution')
plt.xlabel('Items per hour')
plt.ylabel('Frequency')
plt.show()
```

```python
df['accuracy'] = 1 - df['errors'] / df['orders_completed'].replace(0, pd.NA)
sns.boxplot(data=df, x='shift', y='accuracy')
plt.title('Order accuracy by shift')
plt.show()
```

```python
daily = pd.read_csv('../data/daily_kpis.csv', parse_dates=['date'])
daily.plot(x='date', y='avg_pick_rate', title='Daily pick rate trend')
plt.tight_layout()
plt.show()
```
